# Premissas
* Aceitar cachorro;
* Pelo menos 2 quartos;
* Pelo menos 90 m²;
* Não pode ser no térreo;
* Tem que ser na cidade do Porto.


```python
premissas = premissas[(premissas["city"]=="Porto") & (premissas["area"]>=90)]
premissas
```

```python
# Boxplot 
sns.boxplot(data=premissas,x='rooms', y='total')
```
